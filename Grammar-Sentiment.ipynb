# Creating Textual Characteristics (Experiment)

## Grammar

```python
tool = language_tool_python.LanguageTool('en-US')
```

```python
mistakes = []
for i in range(0, len(data)):
    text = data.loc[i, "Text"]
    matches = tool.check(text)
    matches = [match for match in matches if match.ruleId != 'MORFOLOGIK_RULE_EN_US']
    matches = [match for match in matches if match.ruleId != 'CONSECUTIVE_SPACES']
    matches = [match for match in matches if match.ruleId != 'WHITESPACE_RULE']
    mistakes.append(matches)
    data.loc[i, "Grammar"] = len(matches)
```

```python
# Initialize an empty list to store the extracted information
mistakes_data = []

# Iterate over each sublist of mistakes
for sublist in mistakes:
    # Initialize an empty dictionary for each sublist
    sublist_data = {}
    # Extract information for each match in the sublist
    for match in sublist:
        # Extract relevant information from each match
        ruleId = match.ruleId
        message = match.message
        replacements = match.replacements
        offsetInContext = match.offsetInContext
        context = match.context
        offset = match.offset
        errorLength = match.errorLength
        category = match.category
        ruleIssueType = match.ruleIssueType
        sentence = match.sentence
        # Store the extracted information in a dictionary
        match_data = {
            'ruleId': ruleId,
            'message': message,
            'replacements': replacements,
            'offsetInContext': offsetInContext,
            'context': context,
            'offset': offset,
            'errorLength': errorLength,
            'category': category,
            'ruleIssueType': ruleIssueType,
            'sentence': sentence
        }
        # Append the match data to the sublist data
        sublist_data.update(match_data)
    # Append the sublist data to the overall list
    mistakes_data.append(sublist_data)

# Convert the list of dictionaries to a DataFrame
mistakes_df = pd.DataFrame(mistakes_data)
```

## Sentiment

```python
from textblob import TextBlob

sentiment_scores = []

for i in range(0, len(data)):
    text = data.loc[i, "Text"]
    res = TextBlob(text)
    sentiment_scores.append(res.sentiment.polarity)
    data.loc[i, "Sentiment"] = res.sentiment.polarity
```

# Exporting

```python
data.to_csv("experiment_grammar_sentiment.csv", index=False)
```
